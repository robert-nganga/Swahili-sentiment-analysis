```python
#given a list of file names, organize the data into a list
#Process the content of the file line by line and put it into an array.
def organize_data(filenames): # the filenames come in array form.
    processed_data = [] #The array that hold the content of the files
    for filename in filenames: #take the file one by one. One file at a time.
        with open(filename, "r") as file: #Open the file
            for line in file.read().split('\n'): #Split the content of the file line by line
                if len(line.split("\t")) == 2 and line.split("\t")[1] != "":
                    processed_data.append(line.split("\t")) #add each line into the array
        
    return processed_data
```

```python
#Add the polarity of each sentence as either positive(1) or negative(0). The default is positive
def load_data(filename, polarity = 1): #given the name of the file, append polarity at the end of the sentence as either positive or negative.
    data = [] #
    with open(filename, "r", encoding = 'utf8',errors = 'ignore') as file: #Read the content of the file
        #data = [[line, polarity] for line in file.read().split('\n') ]
        for line in file.read().split('\n'): #read a line from the file
            data.append([line, polarity]) #add line and polarity into the array.
    return data
```

```python
#used to evaluate the model to pick the best one
#input are the sentences to train
#output are the correct values to confirm from
#model is the model to evaluate 
#n_folds is the number of times to iterate through the model. To train the model using the same model how many time.

def evaluate_model(inputs, outputs, model, n_folds):
    inputs_copy = np.array(inputs) #use a copy of the array to avoid messing up the original array. this is un
    outputs_copy = np.array(outputs) #use a copy of the array. unnecessary.
    kf = KFold(n_splits = n_folds) #define the split. How to split the dataset to use in training the model.
    scores = list() #list to hold the scores of a model. REsult of how the model performed
    
    for train_index, test_index in kf.split(inputs): #Split the input eg 50, 50, 50,; train with some and test with some, Train with 2 bunch (100 sentences) test with the rest (50)
        inputs_train, inputs_test = inputs_copy[train_index], inputs_copy[test_index]
        outputs_train, outputs_test = outputs_copy[train_index], outputs_copy[test_index]
        model.train(inputs_train, outputs_train)
        predictions = model.predict(inputs_test)
        accuracy = accuracy_score(outputs_test,predictions)
        scores.append(accuracy)
    return scores
```
